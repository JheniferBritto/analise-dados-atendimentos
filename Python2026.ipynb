```python
import pandas as pd

dados = {
    'Paciente': ['Ana', 'Bruno', 'Carlos', 'Daniel', 'Ana', 'Bruno', 'Carlos', 'Ana', 'Marcos'],

    'Data Atendimento': [
        '10/02/2024', 'erro', None, '15/03/2024',
        '32/13/2024', '10-02-2024', '2024/02/10',
        None, '11/02/2024'
    ],

    'Quantidade': [1, '2', None, 100, 2, 'erro', 4, 3, 200],

    'Cidade': ['SP', 'RJ', None, 'SP', '???', 'RJ', 'RJ', 'SP', 'BA'],

    'valor_pago': [100, 250, 150, 750, 100, 400, 320, 200, 500],
    'especialidades':[
    "Triagem", "Teleconsulta", "Sessão Grupo",
    "Home Care", "Avaliação", "Manutenção",
    "Especialista Pleno", "Especialista Sênior", "Alta Complexidade"
]

}

df = pd.DataFrame(dados)
```
